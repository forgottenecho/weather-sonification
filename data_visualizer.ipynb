```python
# imports
import pandas
import matplotlib.pyplot as plt
from midiutil import MIDIFile
```

```python
# load up the data
nj_data = pandas.read_csv('nj_weather_data.csv')
la_data = pandas.read_csv('la_weather_data.csv')
```

```python
# more convenient column names
nj_data.rename(columns={'WT03':'THDR'}, inplace=True)
la_data.rename(columns={'WT03':'THDR'}, inplace=True)

# some datavalues are unrecored, replace the NaN with a 0
nj_data.fillna(0, inplace=True)
la_data.fillna(0, inplace=True)
```

```python
# general statistics on our NJ variables
nj_data[['TMAX','PRCP','THDR','SNOW']].describe()
```

```python
# general statistics on our LA variables
la_data[['TMAX','PRCP','THDR','SNOW']].describe()
```

```python
# create plots for our basic variables
fig, axs = plt.subplots(2,4)

# plot New Jersey data
axs[0, 0].plot(nj_data['TMAX'])
axs[0, 0].set_xlabel('Days since Apr 04 2012')
axs[0, 0].set_ylabel('Temperature (F)')
axs[0, 0].set_title('NJ Daily Max Temperature')

axs[0, 1].plot(nj_data['PRCP'])
axs[0, 1].set_xlabel('Days since Apr 04 2012')
axs[0, 1].set_ylabel('Precipitation (in)') # FIXME units
axs[0, 1].set_title('NJ Daily Precip')

axs[0, 2].plot(nj_data['THDR'])
axs[0, 2].set_xlabel('Days since Apr 04 2012')
axs[0, 2].set_ylabel('Yes 1, No 0')
axs[0, 2].set_title('NJ Presence of Thunder')

axs[0, 3].plot(nj_data['SNOW'])
axs[0, 3].set_xlabel('Days since Apr 04 2012')
axs[0, 3].set_ylabel('Snowfall (in)') # FIXME units
axs[0, 3].set_title('NJ Daily Snowfall')

# plot Louisiana data
axs[1, 0].plot(la_data['TMAX'])
axs[1, 0].set_xlabel('Days since Apr 04 2012')
axs[1, 0].set_ylabel('Temperature (F)')
axs[1, 0].set_title('LA Daily Max Temperature')

axs[1, 1].plot(la_data['PRCP'])
axs[1, 1].set_xlabel('Days since Apr 04 2012')
axs[1, 1].set_ylabel('Precipitation (in)') # FIXME units
axs[1, 1].set_title('LA Daily Precip')

axs[1, 2].plot(la_data['THDR'])
axs[1, 2].set_xlabel('Days since Apr 04 2012')
axs[1, 2].set_ylabel('Yes 1, No 0')
axs[1, 2].set_title('LA Presence of Thunder')

axs[1, 3].plot(la_data['SNOW'])
axs[1, 3].set_xlabel('Days since Apr 04 2012')
axs[1, 3].set_ylabel('Snowfall (in)') # FIXME units
axs[1, 3].set_title('LA Daily Snowfall')

# adjust the subplot spacing
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=2, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=1)
plt.show()
```
